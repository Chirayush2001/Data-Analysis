```python
df['ASSESSABLE VALUE'] = df['ASSESSABLE VALUE'].fillna(0).astype(int)
df['MONTH NAME'] = df['MONTH NAME'].astype(str)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['MONTH NAME'] = pd.to_datetime(df['MONTH NAME'], format='%b-%Y')

df['ASSESSABLE VALUE'] = df['ASSESSABLE VALUE'].round(0).astype(int)

monthly_sales = df.groupby(df['MONTH NAME'].dt.strftime('%b-%Y'))['ASSESSABLE VALUE'].sum()

cumulative_sales = monthly_sales.cumsum()

sorted_data = cumulative_sales.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
bars = plt.bar(sorted_data.index, sorted_data.values, color='y')
plt.xlabel('Month')
plt.ylabel('Cumulative Sales')
plt.title('Cumulative Monthly Sales Trends (Bar Chart)')
plt.grid(axis='y')
plt.xticks(rotation=30)

for bar, value in zip(bars, sorted_data):
    plt.text(bar.get_x() + bar.get_width() / 2, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
filtered_df = df[(df['D.CHANNEL'] == 11) & (df['D.CHANNEL TXT'] == 'Direct')]

filtered_df['CUSTOMER CUMULATIVE PURCHASE'] = filtered_df.groupby('CUSTOMER NAME')['ASSESSABLE VALUE'].cumsum()

top_50_customers = filtered_df.groupby('CUSTOMER NAME')['CUSTOMER CUMULATIVE PURCHASE'].max().nlargest(50)
values = top_50_customers.values

plt.figure(figsize=(16 , 8))
bars = plt.bar(top_50_customers.index, top_50_customers, color='b',label = values)
plt.xlabel('Customer Name')
plt.ylabel('Cumulative Purchase')
plt.xticks(rotation=90)
plt.title('Top 50 Customers where DChannel is Direct')
plt.tight_layout()

for bar, value in zip(bars, top_50_customers):
    plt.text(bar.get_x() + bar.get_width() / 2, value, str(value), ha='center', va='bottom',rotation='vertical', fontsize=8)

plt.show()
```

```python
filtered_df = df[(df['D.CHANNEL'] == 10) & (df['D.CHANNEL TXT'] == 'Dealer / Stockiest')]

filtered_df['CUSTOMER CUMULATIVE PURCHASE'] = filtered_df.groupby('CUSTOMER NAME')['ASSESSABLE VALUE'].cumsum()

top_50_customers = filtered_df.groupby('CUSTOMER NAME')['CUSTOMER CUMULATIVE PURCHASE'].max().nlargest(50)
values = top_50_customers.values

plt.figure(figsize=(16 , 8))
bars = plt.bar(top_50_customers.index, top_50_customers, color='b',label = values)
plt.xlabel('Customer Name')
plt.ylabel('Cumulative Purchase')
plt.xticks(rotation=90)
plt.title('Top 50 Customers where DChannel is Dealer / Stockiest')
plt.tight_layout()

for bar, value in zip(bars, top_50_customers):
    plt.text(bar.get_x() + bar.get_width() / 2, value, str(value), ha='center', va='bottom',rotation='vertical', fontsize=8)

plt.show()
```

```python
df['MIDDLEMAN CUMULATIVE PURCHASE'] = df.groupby('Middle Man Agent Nam')['ASSESSABLE VALUE'].cumsum()

top_50_middlemen = df.groupby('Middle Man Agent Nam')['MIDDLEMAN CUMULATIVE PURCHASE'].max().nlargest(50)

plt.figure(figsize=(16, 6))
bars = plt.bar(top_50_middlemen.index, top_50_middlemen.values, color='b', alpha=0.7, label='Individual Cumulative Purchase')
plt.xlabel('Middle Man Agent')
plt.ylabel('Cumulative Purchase')
plt.xticks(rotation=90)
plt.title('Top 50 Middleman Agents - Cumulative Purchase')
plt.tight_layout()

for bar, value in zip(bars, top_50_middlemen.values):
    plt.text(bar.get_x() + bar.get_width() / 2, value, str(value), ha='center', va='bottom', rotation='vertical', fontsize=8)

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Group the data by 'Profit Center' and sum the 'Assessable Value' for each center
profit_center_revenue = df.groupby('Profit Center')['ASSESSABLE VALUE'].sum().reset_index()

# Sort the data by 'Assessable Value' in descending order
profit_center_revenue = profit_center_revenue.sort_values(by='ASSESSABLE VALUE', ascending=False)

# Create the bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(profit_center_revenue['Profit Center'], profit_center_revenue['ASSESSABLE VALUE'], color='skyblue')
plt.xlabel('Profit Center')
plt.ylabel('Assessable Value')
plt.title('Assessable Value by Profit Center')
plt.grid(axis='y')
plt.xticks(rotation=45)

# Add value annotations on top of each bar
for bar, value in zip(bars, profit_center_revenue['ASSESSABLE VALUE']):
    plt.text(bar.get_x() + bar.get_width() / 2, value, str(value), ha='center', va='bottom')

plt.show()
```
